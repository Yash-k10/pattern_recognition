```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances, cosine_similarity

df = pd.read_csv("customers.csv")

df.columns = df.columns.str.strip()
df.head()
```

```python
euclidean_dist = euclidean_distances(features)

euclidean_df = pd.DataFrame(euclidean_dist)
euclidean_df
```

```python
manhattan_dist = manhattan_distances(features)

manhattan_df = pd.DataFrame(manhattan_dist)
manhattan_df
```

```python
cos_sim = cosine_similarity(features)

cosine_df = pd.DataFrame(cos_sim)
cosine_df
```

```python
df.index = [f"C{i}" for i in range(len(df))]
cosine_df.index = df.index
cosine_df.columns = df.index
cosine_df
```

```python
def most_similar(customer, matrix):
    idx = df.index.get_loc(customer)
    scores = matrix[idx]

    sorted_idx = np.argsort(scores)[::-1]

    for i in sorted_idx:
        if i != idx:
            return df.index[i]

print("Most similar to C0:", most_similar("C0", cos_sim))
```
