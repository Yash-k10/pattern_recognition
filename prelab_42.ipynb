```python
from sklearn.datasets import load_iris

iris = load_iris()

print(iris.keys())

import pandas as pd

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df['target'] = iris.target

df.head()

print(df.shape)

print(df.info())

print(df.describe())
print(df.isnull().sum())

plt.figure(figsize=(8,5))
plt.scatter(
    df['sepal length (cm)'],
    df['petal length (cm)']
)

plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Iris Dataset Scatter Plot")
plt.show()
```
